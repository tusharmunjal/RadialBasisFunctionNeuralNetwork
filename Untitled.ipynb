```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class RBFNN:
    def __init__(self, num_centers, gamma=1.0):
        self.num_centers = num_centers
        self.gamma = gamma

    def fit(self, X, y):
        # Step 1: Cluster the input data using KMeans
        kmeans = KMeans(n_clusters=self.num_centers, random_state=42)
        kmeans.fit(X)
        self.centers = kmeans.cluster_centers_

        # Step 2: Calculate the spread parameter for each center
        distances = euclidean_distances(X, self.centers)
        self.sigma = np.mean(np.max(distances, axis=1))

        # Step 3: Compute the hidden layer activations
        self.phi = np.exp(-self.gamma * (distances / self.sigma) ** 2)

        # Step 4: Solve for the output weights using pseudo-inverse
        self.output_weights = np.linalg.pinv(self.phi) @ y

    def predict(self, X):
        distances = euclidean_distances(X, self.centers)
        phi = np.exp(-self.gamma * (distances / self.sigma) ** 2)
        y_pred = phi @ self.output_weights
        return y_pred

# Generate synthetic dataset
X = np.random.rand(100, 2)  # 100 samples, 2 features
y = (X[:, 0] + X[:, 1] > 1).astype(int)  # Binary classification problem

# One-hot encode the target labels
encoder = OneHotEncoder()
y_onehot = encoder.fit_transform(y.reshape(-1, 1)).toarray()

# Split dataset into training and testing sets
X_train, X_test = X[:80], X[80:]
y_train, y_test = y_onehot[:80], y_onehot[80:]

# Initialize and train RBFNN
num_centers = 10
rbfnn = RBFNN(num_centers=num_centers)
rbfnn.fit(X_train, y_train)

# Make predictions
y_pred = rbfnn.predict(X_test)
y_pred_binary = np.argmax(y_pred, axis=1)

# Calculate accuracy
accuracy = accuracy_score(np.argmax(y_test, axis=1), y_pred_binary)
print("Accuracy:", accuracy)
```
